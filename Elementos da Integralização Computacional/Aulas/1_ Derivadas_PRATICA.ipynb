# O que são derivadas?

Uma derivada é um conceito fundamental da matemática que `descreve a taxa de variação de uma função em relação à mudança de sua variável independente.`<br>
    <b>Em termos mais simples, a derivada 'MEDE' como uma função "MUDA" à medida que sua entrada (variável independente) muda.


Matematicamente, a derivada de uma função $$ f(x) $$ em relação a $$x$$ é denotada como $$f'(x) $$ ou $$ \frac{df}{dx}$$. 

Matematicamente, a derivada é calculada como o `limite do quociente diferencial conforme o intervalo de mudança na variável independente se aproxima de zero`:

$$f'(x) = \lim_{{h \to 0}} \frac{f(x + h) - f(x)}{h} $$

O valor da derivada em um ponto $x$ representa a inclinação da reta tangente à curva da função nesse ponto. Se a derivada for:<br>
> `Positiva`, a função está aumentando no ponto.

> `Negativa`, a função está diminuindo.

> `Zero`, a função tem um ponto crítico onde sua inclinação é horizontal.

Para entender o conceito de derivadas de forma mais profunda, é importante estudar cálculo diferencial e suas aplicações. Elas desempenham um papel crucial na compreensão do comportamento de funções e na resolução de problemas do mundo real que envolvem mudanças e variações.

# Representação gráfica

- **Sem a biblioteca `SYMPY`**

```python
#pip install matplotlib numpy
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função e sua derivada
def funcao(x):
    return x**2 + 2*x + 1

def derivada(x):
    return 2*x + 2

# Intervalo de x
x = np.linspace(-5, 5, 400)

# Calculando os valores da função e da derivada em cada ponto
y_funcao = funcao(x)
y_derivada = derivada(x)

# Plotando os gráficos
plt.figure(figsize=(10, 6))
plt.plot(x, y_funcao, label='Função: $f(x) = x^2 + 2x + 1$')
plt.plot(x, y_derivada, label='Derivada: $f\'(x) = 2x + 2$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da Função e sua Derivada')
plt.legend()
plt.grid(True)
plt.show()
```

Neste exemplo, a função $$ f(x) = x^2 + 2x + 1 $$ <br> foi definida, juntamente com sua derivada  $$f'(x) = 2x + 2 $$ <br> A biblioteca `numpy` é usada para criar um conjunto de valores de $x$ ao longo de um intervalo. Em seguida, os valores correspondentes de $y$ para a função e sua derivada são calculados usando essas funções. Finalmente, os gráficos são plotados usando a `matplotlib`.

- **Com a biblioteca `SYMPY`**

Você pode representar graficamente as derivadas de funções usando Python e as bibliotecas `SymPy` e `matplotlib`. Aqui está um exemplo de como fazer isso:

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definindo a variável simbólica
x = sp.symbols('x')

# Função original
funcao = x**3 - 2*x**2 + x

# Calculando a derivada
derivada = sp.diff(funcao, x)

# Convertendo a função e a derivada para funções lambdas
f_lambdified = sp.lambdify(x, funcao, "numpy")
f_prime_lambdified = sp.lambdify(x, derivada, "numpy")

# Intervalo de x para plotagem
x_vals = np.linspace(-2, 3, 400)

# Calculando os valores da função e da derivada nos pontos x_vals
f_vals = f_lambdified(x_vals)
f_prime_vals = f_prime_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Gráfico da função original
plt.plot(x_vals, f_vals, label='Função: $f(x) = x^3 - 2x^2 + x$')

# Gráfico da derivada
plt.plot(x_vals, f_prime_vals, label='Derivada: $f\'(x) = 3x^2 - 4x + 1$', linestyle='dashed')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Função e sua Derivada')
plt.legend()
plt.grid(True)
plt.show()
```

Neste exemplo, a função original é $$ f(x) = x^3 - 2x^2 + x $$ e sua derivada é $$ f'(x) = 3x^2 - 4x + 1 $$ <br> O código calcula os valores da função e da derivada em um intervalo de $ x $ e, em seguida, traça ambos os gráficos usando a biblioteca `matplotlib`.

> Lembre-se de que, neste caso, a derivada é representada como uma linha pontilhada para distingui-la da função original. Isso ajuda a visualizar como a inclinação da função original muda em diferentes pontos.

# Interpretação das derivadas

> As derivadas podem ser interpretada de várias maneiras:

## **Taxa de Mudança Instantânea:** 
A derivada de uma função em um ponto específico $x$ representa a taxa de mudança da função nesse ponto. Em outras palavras, é a inclinação da reta tangente à curva da função nesse ponto.

A taxa de mudança instantânea de uma função é representada pela derivada da função em relação à variável independente. Vou mostrar como calcular e visualizar a taxa de mudança instantânea usando Python, as bibliotecas `SymPy` e `matplotlib`.

Neste exemplo, usaremos a função $$f(x) = x^2 $$ para calcular a taxa de mudança instantânea em diferentes pontos e plotar o gráfico da função e sua derivada, que representa a taxa de mudança.

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definindo a variável simbólica
x = sp.symbols('x')

# Definindo a função
funcao = x**2

# Calculando a derivada (taxa de mudança)
taxa_mudanca = sp.diff(funcao, x)

# Convertendo a função e a taxa de mudança para funções lambdas
f_lambdified = sp.lambdify(x, funcao, "numpy")
taxa_mudanca_lambdified = sp.lambdify(x, taxa_mudanca, "numpy")

# Intervalo de x para plotagem
x_vals = np.linspace(-2, 2, 400)

# Calculando os valores da função e da taxa de mudança nos pontos x_vals
f_vals = f_lambdified(x_vals)
taxa_vals = taxa_mudanca_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Gráfico da função
plt.plot(x_vals, f_vals, label='Função: $f(x) = x^2$')

# Gráfico da taxa de mudança (derivada)
plt.plot(x_vals, taxa_vals, label='Taxa de Mudança (Derivada)', linestyle='dashed')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Função e sua Taxa de Mudança (Derivada)')
plt.legend()
plt.grid(True)
plt.show()
```



Neste exemplo, o gráfico da função $$f(x) = x^2$$ é plotado junto com o gráfico da derivada $$ f'(x) = 2x $$, que representa a taxa de mudança instantânea em cada ponto. A derivada indica a inclinação da reta tangente à curva da função original em cada ponto.

## **Velocidade e Aceleração:**
Se a função representa uma quantidade física em função do tempo, sua derivada pode representar a velocidade (primeira derivada) e a aceleração (segunda derivada).

Neste exemplo, usaremos uma função para representar a `posição de uma partícula ao longo do tempo e, em seguida, calcularemos a velocidade (derivada da posição) e a aceleração (derivada da velocidade).`


```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definindo a variável simbólica
t = sp.symbols('t')

# Definindo a função de posição
posicao = 2*t**2 + 3*t + 1

# Calculando a derivada da posição para obter a velocidade
velocidade = sp.diff(posicao, t)

# Calculando a derivada da velocidade para obter a aceleração
aceleracao = sp.diff(velocidade, t)

# Convertendo as funções para funções lambdas
posicao_lambdified = sp.lambdify(t, posicao, "numpy")
velocidade_lambdified = sp.lambdify(t, velocidade, "numpy")
aceleracao_lambdified = sp.lambdify(t, aceleracao, "numpy")

# Intervalo de tempo para plotagem
t_vals = np.linspace(0, 5, 400)

# Calculando os valores da posição, velocidade e aceleração nos pontos t_vals
posicao_vals = posicao_lambdified(t_vals)
velocidade_vals = velocidade_lambdified(t_vals)
aceleracao_vals = aceleracao_lambdified(t_vals)


## CUIDADO!!! Quando o produto da derivada é uma constante iremos ter um único digito o que dificulta a representação gráfica, portanto, criamos um vetor de n-posições igual ao vetor de tempo e deppois é multiplicado pela constante, assim poderemos gráficar
# Criando um vetor - Aceleração constante
if type(aceleracao_vals) == int:
    size = len(t_vals)
    vetor = np.ones(size)
    aceleracao_vals = vetor * aceleracao_vals

# Criando o gráfico
plt.figure(figsize=(12, 8))

# Gráfico da posição
plt.subplot(3, 1, 1)
plt.plot(t_vals, posicao_vals, label='Posição: $s(t) = 2t^2 + 3t + 1$')
plt.ylabel('Posição')
plt.title('Posição, Velocidade e Aceleração')
plt.legend()

# Gráfico da velocidade
plt.subplot(3, 1, 2)
plt.plot(t_vals, velocidade_vals, label='Velocidade: $v(t) = 4t + 3$')
plt.ylabel('Velocidade')
plt.legend()

# Gráfico da aceleração
plt.subplot(3, 1, 3)
plt.plot(t_vals, aceleracao_vals, label='Aceleração: $a(t) = 4$')
plt.xlabel('Tempo')
plt.ylabel('Aceleração')
plt.legend()

plt.tight_layout()
plt.show()
```


Neste exemplo, calculamos a posição, velocidade e aceleração de uma partícula ao longo do tempo e plotamos os gráficos correspondentes. A posição é dada pela função $$ s(t) = 2t^2 + 3t + 1 $$, a velocidade é a derivada da posição $$ v(t) = 4t + 3 $$, e a aceleração é a derivada da velocidade $$ a(t) = 4 $$. Observe que a aceleração é constante neste caso.

## **Taxa de Crescimento:** 
Se a função representa uma quantidade em relação a outra quantidade (por exemplo, população em relação ao tempo), sua derivada pode representar a taxa de crescimento ou decrescimento.

Para calcular a `taxa de crescimento` de uma função usando derivadas em Python, você pode seguir os passos que mostro abaixo. Vou usar o exemplo de uma função que `representa a população de uma cidade em relação ao tempo`.

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definindo a variável simbólica
t = sp.symbols('t')

# Função que representa a população em relação ao tempo (exemplo)
populacao = 1000 * sp.exp(0.05*t)

# Calculando a taxa de crescimento (derivada)
taxa_crescimento = sp.diff(populacao, t)

# Convertendo a função e a taxa de crescimento para funções lambdas
populacao_lambdified = sp.lambdify(t, populacao, "numpy")
taxa_crescimento_lambdified = sp.lambdify(t, taxa_crescimento, "numpy")

# Intervalo de tempo para plotagem
t_vals = np.linspace(0, 20, 400)

# Calculando os valores da população e da taxa de crescimento nos pontos t_vals
populacao_vals = populacao_lambdified(t_vals)
taxa_crescimento_vals = taxa_crescimento_lambdified(t_vals)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Gráfico da população
plt.subplot(2, 1, 1)
plt.plot(t_vals, populacao_vals, label='População: $P(t) = 1000e^{0.05t}$')
plt.ylabel('População')
plt.title('População e Taxa de Crescimento')
plt.legend()

# Gráfico da taxa de crescimento
plt.subplot(2, 1, 2)
plt.plot(t_vals, taxa_crescimento_vals, label='Taxa de Crescimento: $P\'(t) = 50e^{0.05t}$', linestyle='dashed')
plt.xlabel('Tempo')
plt.ylabel('Taxa de Crescimento')
plt.legend()

plt.tight_layout()
plt.show()
```

Neste exemplo, `a função representa a população de uma cidade em relação ao tempo`, e a taxa de crescimento é calculada como a derivada da função. O gráfico mostra como a população cresce ao longo do tempo e como a taxa de crescimento varia. A taxa de crescimento é representada pela derivada da função original.

## **Taxa de Variação:** 

A derivada também pode ser interpretada como a taxa de variação da função em relação à mudança na variável independente. Por exemplo, em uma função de posição em relação ao tempo, a derivada é a velocidade, que mede a taxa de variação da posição em relação ao tempo.

Para calcular e representar a taxa de variação de uma função usando Python, você pode seguir uma abordagem semelhante à taxa de crescimento que mencionei anteriormente. Vou usar um exemplo onde calcularemos `a taxa de variação de uma função que representa a distância percorrida por um carro em relação ao tempo`.


```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definindo a variável simbólica
t = sp.symbols('t')

# Função que representa a distância percorrida pelo carro em relação ao tempo
distancia = 20*t**2 + 10*t

# Calculando a taxa de variação (derivada)
taxa_variacao = sp.diff(distancia, t)

# Convertendo a função e a taxa de variação para funções lambdas
distancia_lambdified = sp.lambdify(t, distancia, "numpy")
taxa_variacao_lambdified = sp.lambdify(t, taxa_variacao, "numpy")

# Intervalo de tempo para plotagem
t_vals = np.linspace(0, 5, 400)

# Calculando os valores da distância e da taxa de variação nos pontos t_vals
distancia_vals = distancia_lambdified(t_vals)
taxa_variacao_vals = taxa_variacao_lambdified(t_vals)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Gráfico da distância percorrida pelo carro
plt.subplot(2, 1, 1)
plt.plot(t_vals, distancia_vals, label='Distância: $d(t) = 20t^2 + 10t$')
plt.ylabel('Distância')
plt.title('Distância e Taxa de Variação')
plt.legend()

# Gráfico da taxa de variação
plt.subplot(2, 1, 2)
plt.plot(t_vals, taxa_variacao_vals, label='Taxa de Variação: $d\'(t) = 40t + 10$', linestyle='dashed')
plt.xlabel('Tempo')
plt.ylabel('Taxa de Variação')
plt.legend()

plt.tight_layout()
plt.show()
```


Neste exemplo, a `função` representa a **distância percorrida por um carro em relação ao tempo**, e a `taxa de variação` é calculada como a **derivada da função**. O gráfico mostra como a distância percorrida pelo carro muda ao longo do tempo e como a taxa de variação varia. A taxa de variação é representada pela derivada da função original.

# Formulas das derivadas


Aqui estão algumas fórmulas úteis para calcular derivadas de funções comuns. Lembre-se de que $f'(x)$ denota a derivada de $f(x)$ em relação a $x$

`REGRAS BÁSICAS`

- **Variável independente:**<br>

$ \frac{d}{dx} (x) = 1 $

- **Potência:**<br>

$ \frac{d}{dx} x^n = n \cdot x^{n-1} $
   
- **Constante:**<br>

$ \frac{d}{dx} c = 0 $ onde $c $ é uma constante

- **Multiplicação por Constante:**<br>

$ \frac{d}{dx} (c \cdot f(x)) = c \cdot \frac{d}{dx} f(x) $

- **Soma/Diferença:**<br>

$ \frac{d}{dx} (f(x) \pm g(x)) = \frac{d}{dx} f(x) \pm \frac{d}{dx} g(x) $

`REGRA DO PRODUTO E COCIENTE`
- **Regra do Produto:**<br>

$ \frac{d}{dx} (uv) = u'v + uv' $
   
- **Regra do Quociente:** <br>

$ \frac{d}{dx} \left(\frac{u}{v}\right) = \frac{u'v - uv'}{v^2} $
      
`EXPONENCIAL E LOGARITMICA`

- **Exponencial:**<br>
$ \frac{d}{dx} e^x = e^x $

- **Logaritmo Natural:**<br>

$ \frac{d}{dx} \ln(x) = \frac{1}{x} $


`Regra da Cadeia:`<br>

$ \frac{d}{dx} (f(g(x))) = f'(g(x)) \cdot g'(x) $

`FUNÇÕES TRIGONOMETRICAS`

- **Função Seno:**<br>

$ \frac{d}{dx} \sin(x) = \cos(x) $

- **Função Cosseno:**<br>

$ \frac{d}{dx} \cos(x) = -\sin(x) $

- **Função Tangente:**<br>

$ \frac{d}{dx} \tan(x) = \sec^2(x) $

- **Função Arcseno (Inversa do Seno):**<br>

$ \frac{d}{dx} \arcsin(x) = \frac{1}{\sqrt{1-x^2}} $

- **Função Arccosseno (Inversa do Cosseno):**<br>

$ \frac{d}{dx} \arccos(x) = -\frac{1}{\sqrt{1-x^2}} $

- **Função Arctangente (Inversa da Tangente):**<br>

$ \frac{d}{dx} \arctan(x) = \frac{1}{1+x^2} $


> Estas são apenas algumas das fórmulas mais comuns para calcular derivadas. Lembre-se de que as derivadas podem ser aplicadas em funções mais complexas e em combinação com outras regras. Além disso, a prática constante é essencial para desenvolver habilidades sólidas em cálculo diferencial.

## Exemplos práticos

### Cálculo de Velocidade a partir de uma Função de Posição:

> Suponha que temos uma função de posição em relação ao tempo e queremos calcular a velocidade instantânea:

```python
t = sp.symbols('t')
s = 2*t**2 + 3*t + 1  # Função de posição

v = sp.diff(s, t)  # Derivada da posição em relação ao tempo é a velocidade
print("Velocidade:", v)
```

### Cálculo de Taxa de Crescimento:

> Digamos que temos uma função que modela o crescimento de uma população e queremos calcular a taxa de crescimento:

```python
p = sp.Function('p')(t)  # Função que modela a população
p = 100*sp.exp(0.1*t)  # Exemplo de função populacional

taxa_crescimento = sp.diff(p, t)  # Derivada da população em relação ao tempo é a taxa de crescimento
print("Taxa de Crescimento:", taxa_crescimento)
```

### Cálculo de Elasticidade:

> Em economia, a elasticidade é uma medida da sensibilidade de uma variável em relação a outra. Vamos calcular a elasticidade de um preço em relação à quantidade demandada:

```python
Q, P = sp.symbols('Q P')
Q = 100 - 2*P  # Quantidade demandada como função do preço

elasticidade = sp.diff(Q, P) * (P/Q)
print("Elasticidade de Preço em relação à Quantidade:", elasticidade)
```

2. **Paul's Online Math Notes:** O professor Paul Dawkins oferece uma série de notas de matemática, incluindo tutoriais sobre cálculo. Ele também fornece uma seção de exercícios com soluções detalhadas.

   Site: [Paul's Online Math Notes - Calculus I](http://tutorial.math.lamar.edu/Classes/CalcI/CalcI.aspx)

3. **Calcworkshop:** O Calcworkshop é um site que oferece aulas em vídeo e exercícios em formato de quiz para ajudar você a praticar cálculo, incluindo derivadas.

   Site: [Calcworkshop - Derivatives](https://www.calcworkshop.com/derivatives/)

4. **Wolfram Alpha:** Embora seja mais conhecido por sua capacidade de resolver equações, o Wolfram Alpha também oferece prática interativa de cálculo, incluindo derivadas.

   Site: [Wolfram Alpha - Practice Calculus](https://www.wolframalpha.com/problem-generator/calculus/)

5. **Soluções de Livros de Texto:** Muitos livros de cálculo, como os de James Stewart, Guidorizzi e outros autores, possuem seções de exercícios de derivadas com respostas no final do livro. Usar esses exercícios pode ser uma ótima maneira de praticar.

Lembre-se de que a prática constante é fundamental para o aprendizado eficaz. Reserve um tempo regular para resolver exercícios e trabalhar com derivadas para melhorar suas habilidades ao longo do tempo.
