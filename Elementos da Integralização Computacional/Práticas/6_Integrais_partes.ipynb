# Preparação inicial

```python
# Imports
import sympy as sp
from sympy import E, sin, asin, atan, cos, ln, pi
from sympy import init_printing
import numpy as np
import matplotlib.pyplot as plt

# Ativando o pretty printing
init_printing(order='none')
```

## Integrais indefinidas

## Questão 1
$\int x^2e^xdx$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')
C = sp.Symbol('C')

# Definindo a função
f = x**2*E**x

# Integral indefinida
F = sp.integrate(f, x)
F + C
```

```python
# No primeiro subgráfico, iremos plotar a área de 0 -> 2
# então precisamos calcular o valor da integral definida nesse intervalo
area = sp.integrate(f, (x, 0, 2))
area
```

```python
# Plotando os gráficos

# Convertendo em funções lambda
f_lambdified = sp.lambdify(x, f, 'numpy')
F_lambdified = sp.lambdify(x, F, 'numpy')

# x_vals
x_vals = np.linspace(0, 2.5, 400)

# f_vals e F_vals
f_vals = f_lambdified(x_vals)
F_vals = F_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(12, 8))

# Subgráfico 1: f(x) e F(x) como área sob curva
# Plotando a área
plt.subplot(2, 1, 1)
plt.plot(x_vals, f_vals, label=r'$f(x) = x^2e^x$', color='orange')
plt.fill_between(x_vals, f_vals, where=(x_vals>=0)&(x_vals<=2), label=r'A = $2e^2-2$ u.a.', color='orange', alpha=0.3)
# Limitando visualmente o gráfico
plt.xlim(0, 2.5)
plt.ylim(0, 40)
# plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(2, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Grade e título
plt.grid(linestyle='--', alpha=0.3)
plt.title('f(x) e F(x) como área sob curva')
plt.legend()

# Subgráfico 2: Comparação entre f(x) e F(x)
plt.subplot(2, 1, 2)
plt.plot(x_vals, f_vals, label=r'$f(x) = x^2e^x$', color='blue')
plt.plot(x_vals, F_vals, label=r'$F(x) = x^2e^x-2xe^x-2e^x$', color='red')
# Limitando visualmente o gráfico
plt.xlim(0, 2.5)
plt.ylim(0, 40)
# Grade e título
plt.grid(linestyle='--', alpha=0.3)
plt.title('Comparação entre f(x) e F(x)')
plt.legend()
plt.show()
```

## Questão 3
$\int \ln{x^2}dx$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')
C = sp.Symbol('C')

# Definindo a função
f = ln(x**2)

# Integral indefinida
F = sp.integrate(f, x)
F + C
```

```python
# Iremos plotar a área de entre os pontos de interseção
# então precisamos calcular o valor da integral definida nesse intervalo

# Calculando os pontos de interseção
eq = sp.Eq(f, F)

# O sympy não resolve essa equação simbolicamente, então usamos o sp.nsolve,
# passando um chute de onde encontraremos o ponto

# Ponto 1 (0.446432978428280, -0.806465994236327)
ponto_1_x = float(sp.nsolve(eq, 0.4))
ponto_1_y = float(sp.log(ponto_1_x**2))

# Ponto 2 (3.85733482594938, 1.34997648540113)
ponto_2_x = float(sp.nsolve(eq, 3.5))
ponto_2_y = float(sp.log(ponto_2_x**2))

# Calculando a integral definida entre os pontos
# A área entre duas curvas f(x) e g(x) é \int_a^b |f(x)| - |g(x)| dx
# f fica acima de F (f > F), então:
expr = f - F
area = sp.integrate(expr, (x, ponto_1_x, ponto_2_x))

print(f"A = {area} u.a.\n")
print(f"P1 = ({ponto_1_x}, {ponto_1_y})\nP2 = ({ponto_2_x}, {ponto_2_y})")
```

```python
# Plotando os gráficos

# Gráfico 1: Comparação entre f(x) e F(x)
# Convertendo em funções lambda
f_lambdified = sp.lambdify(x, f, 'numpy')
F_lambdified = sp.lambdify(x, F, 'numpy')
# x_vals
x_vals = np.linspace(-4, 4, 800)
# f_vals e F_vals
f_vals = f_lambdified(x_vals)
F_vals = F_lambdified(x_vals)
# Criando o gráfico
plt.figure(figsize=(12, 8))
# Plotando o gráfico
plt.plot(x_vals, f_vals, label=r'$f(x) = \ln{x^2}$', color='blue')
plt.plot(x_vals, F_vals, label=r'$F(x) = 2x(\ln{x}-1)$', color='red')
# Limitando visualmente
plt.xlim(-4, 4)
plt.ylim(-2, 3)
# Mostrando os pontos que delimitam a área plotada no gráfico 2
plt.scatter([ponto_1_x, ponto_2_x], [ponto_1_y, ponto_2_y], color='red', zorder=3)
plt.text(ponto_1_x+0.1, ponto_1_y, r"$(0.44, -1.61)$", bbox=dict(boxstyle='round', facecolor='#ffc042', alpha=0.7))
plt.text(ponto_2_x-0.9, ponto_2_y, r"$(3.85, 2.69)$", bbox=dict(boxstyle='round', facecolor='#ffc042', alpha=0.7))
# Eixos x e y, grade e título
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.grid(linestyle='--', alpha=0.3)
plt.title('Comparação entre f(x) e F(x)')
plt.legend()

# Gráfico 2: Área entre curvas de f(x) e F(x)
# x_vals
x_vals = np.linspace(ponto_1_x-0.1, ponto_2_x+0.1, 800)
# f_vals e F_vals
f_vals = f_lambdified(x_vals)
F_vals = F_lambdified(x_vals)
# Criando o gráfico
plt.figure(figsize=(12, 8))
# Plotando o gráfico
plt.plot(x_vals, f_vals, label=r'$f(x) = \ln{x^2}$', color='orange')
plt.plot(x_vals, F_vals, label=r'$F(x) = 2x(\ln{x}-1)$', color='orange')
# x_fill, f_fill, F_fill
x_fill = np.linspace(ponto_1_x, ponto_2_x, 400)
f_fill = f_lambdified(x_fill)
F_fill = F_lambdified(x_fill)
plt.fill_between(x_fill, f_fill, F_fill, label=r'A = $\approx6.0854$ u.a.', color='#00b3ff', alpha=0.5)
# Plotando a legenda dos pontos
plt.scatter([ponto_1_x, ponto_2_x], [ponto_1_y, ponto_2_y], color='red', zorder=3)
plt.text(ponto_1_x, ponto_1_y-0.3, r"$(0.44, -1.61)$", bbox=dict(boxstyle='round', facecolor='#ffc042', alpha=0.7))
plt.text(ponto_2_x, ponto_2_y-0.3, r"$(3.85, 2.69)$", bbox=dict(boxstyle='round', facecolor='#ffc042', alpha=0.7))
# Eixo x, grade e título
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.grid(linestyle='--', alpha=0.3)
plt.title('f(x) e F(x) como área sob curva')
plt.legend()

plt.show()
```

```python
# Plotando os gráficos

# Convertendo em funções lambda
f_lambdified = sp.lambdify(x, f, 'numpy')
F_lambdified = sp.lambdify(x, F, 'numpy')

# x_vals
x_vals = np.linspace(-3, 1.5, 800)

# f_vals e F_vals
f_vals = f_lambdified(x_vals)
F_vals = F_lambdified(x_vals)

# Criando o gráfico: Comparação entre f(x) e F(x)
plt.figure(figsize=(12, 8))
plt.plot(x_vals, f_vals, label=r'$f(x) = xe^x$', color='blue')
plt.plot(x_vals, F_vals, label=r'$F(x) = e^x(x-1)$', color='red')
# Limitando visualmente o gráfico
plt.xlim(-3, 1.5)
plt.ylim(-1, 3)
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Grade e título
plt.grid(linestyle='--', alpha=0.3)
plt.title('Comparação entre f(x) e F(x)')
plt.legend()

# Criando o 2º gráfico
plt.figure(figsize=(12, 8))

# Subgráfico 1: f(x) e F(x) como área sob curva
# Plotando a área
plt.subplot(1, 2, 1)
plt.plot(x_vals, f_vals, label=r'$f(x) = xe^x$', color='orange')
plt.fill_between(x_vals, f_vals, where=(x_vals>=0)&(x_vals<=1), label=r'A = $1$ u.a.', color='orange', alpha=0.3)
# Limitando visualmente o gráfico
plt.xlim(-1, 1.5)
plt.ylim(-0.5, 3)
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Grade e título
plt.grid(linestyle='--', alpha=0.3)
plt.title('f(x) e F(x) como área sob curva')
plt.legend()

# Subgráfico 2: f(x) e sua integral acumulada
# Definindo a primitiva acumulada
F_acumulada = sp.integrate(f, (x, 0, x))
F_acumulada_lambda = sp.lambdify(x, F_acumulada, 'numpy')
F_acumulada_vals = F_acumulada_lambda(x_vals)
# Plotando o gráfico
plt.subplot(1, 2, 2)
plt.plot(x_vals, f_vals, label=r'$f(x) = xe^x$')
plt.plot(x_vals, F_acumulada_vals, label=r'$F(x) = e^x(x-1)$', linestyle='--')
# Plotando o ponto final
plt.scatter([1], [F_acumulada_lambda(1)], color='red', zorder=3, label=r'$0 \to F(1)=1$')
plt.scatter([0], [0], color='red', zorder=3)
# Limitando visualmente o gráfico
plt.xlim(-1, 1.5)
plt.ylim(-0.5, 3)
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Grade e título
plt.grid(linestyle='--', alpha=0.3)
plt.title("Função e integral acumulada")
plt.legend()

plt.show()
```
