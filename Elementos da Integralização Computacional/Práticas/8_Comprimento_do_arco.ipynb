# Resoluções

```python
# Imports
import sympy as sp
from sympy import sin, cos, pi, log, csc, cot, sqrt
from sympy import init_printing
import numpy as np
import matplotlib.pyplot as plt

# Ativando o pretty printing
init_printing(order='none')
```

# Questões

## Questão 1
$y=5x-2; [-2,2]$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')

# Definindo as funções
y = 5*x - 2

# Área desejada
derivada = sp.diff(y, x)
# derivada
area = sp.integrate(sqrt(1+derivada**2), (x, -2, 2))
area
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
y_lambda = sp.lambdify(x, y, 'numpy')

# x_vals
x_vals = np.linspace(-2, 2, 1000)

# y_vals
y_vals = y_lambda(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label=r'$y=5x-2 [-2,2]$', color="red")
#ax.plot([])

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(2, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(-2, color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
#plt.xlim(-1, 1)
#plt.ylim(-2,2)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$y=5x-2;\:L=4\sqrt{26}u.c.;\:[-2,2]$')
plt.legend()
plt.show()
```

## Questão 2
$y=x^\frac{2}{3}-1;\:[1,2]$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')

# Definindo as funções
y = x**sp.Rational(2, 3)-1

# Área desejada
derivada = sp.diff(y, x)
# derivada

integrando = sp.sqrt(1+derivada**2)
comprimento = sp.integrate(integrando, (x, 1, 2))
comprimento
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
y_lambda = sp.lambdify(x, y, 'numpy')

# x_vals
x_vals = np.linspace(0, 6, 1000)

# y_vals
y_vals = y_lambda(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label=r'$y=x^\frac{2}{3}-1;[1,2]$', color="red")
#ax.plot([])

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(2, color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
plt.xlim(0, 6)
plt.ylim(-1, 4)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$y=x^\frac{2}{3}-1;\:L=\frac{1}{27}[(9*2^\frac{2}{3}+4)^\frac{3}{2};\:[1,2]$')
plt.legend()
plt.show()
```

## Questão 3
$y=\frac{1}{3}(2+x^2)^\frac{3}{2};\:[0,3]$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')

# Definindo as funções
y = (sp.Rational(1, 3))*(2+x**2)**(sp.Rational(3, 2))

# Área desejada
derivada = sp.diff(y, x)
# derivada
area = sp.integrate(sp.sqrt(1+derivada**2), (x, 0, 3))
area.evalf()
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
y_lambda = sp.lambdify(x, y, 'numpy')

# x_vals
x_vals = np.linspace(0, 3, 1000)

# y_vals
y_vals = y_lambda(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label=r'$y=\frac{1}{3}(2+x^2)^\frac{3}{2};\:[0,3]$', color="red")

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(3, color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
#plt.xlim(-1, 1)
#plt.ylim(-2,2)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$y=\frac{1}{3}(2+x^2)^\frac{3}{2};\:[0,3];\:L=12\:u.c.;\:[0,3]$')
plt.legend()
plt.show()
```

## Questão 4
$x^\frac{2}{3}+y^\frac{2}{3}=2^\frac{2}{3};\:[0,3]$

```python
# Definindo as variáveis simbólicas
t = sp.Symbol('t')

# Funções
x = 2*cos(t)**3
y = 2*sin(t)**3

# Integrando
derivada_x = sp.diff(x, t)
derivada_y = sp.diff(y, t)

# Definindo as funções
comprimento = sp.integrate(sp.sqrt(((derivada_x)**2)+((derivada_y)**2)), (t, 0, 2*pi))
comprimento
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
x_lambda = sp.lambdify(t, x, 'numpy')
y_lambda = sp.lambdify(t, y, 'numpy')

# t_vals
t_vals = np.linspace(0, 2*np.pi, 1000)

# y_vals
y_vals = y_lambda(t_vals)
x_vals = x_lambda(t_vals)

# Criando o gráfico
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals, label=r'$x^\frac{2}{3}+y^\frac{2}{3}=2^\frac{2}{3};\:[0,3]$', color="red")

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(2*np.pi, color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
plt.xlim(-2, 2)
plt.ylim(-2, 2)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$x^\frac{2}{3}+y^\frac{2}{3}=2^\frac{2}{3};\:L=4\sqrt{26}u.c.;\:[0,3]$')
plt.legend()
plt.show()
```

## Questão 5
$y=\ln{x};\:[\sqrt{3},\sqrt{8}]$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')

# Definindo as funções
y = log(x)

# Derivada
dy = sp.diff(y, x)
# dy

# Integrando comprimento
integrando = sp.sqrt(1+dy**2)

area = sp.simplify(sp.integrate(integrando, (x, sqrt(3), sqrt(8))))
area
```

```python
area_log = sp.simplify(area.rewrite(sp.log))

sp.pprint(area_log)
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
y_lambda = sp.lambdify(x, y, 'numpy')

# x_vals
x_vals = np.linspace(np.sqrt(3)-0.2, np.sqrt(8)+0.2, 1000)

# y_vals
y_vals = y_lambda(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label=r'$y=\ln{x};\:[\sqrt{3},\sqrt{8}]$', color="red")
#ax.plot([])

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(np.sqrt(3), color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(np.sqrt(8), color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
plt.xlim(1.5, 3)
plt.ylim(0.2, 1.2)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$y=\ln{x};\:L=1+\frac{1}{2}\:\ln{\frac{3}{2}}\:u.c.;\:[\sqrt{3},\sqrt{8}]$')
plt.legend()
plt.show()
```

## Questão 6
$y=1-\ln{\sin{x}};\:[\frac{\pi}{6},\frac{\pi}{4}]$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')

# Definindo as funções
y = 1-log(sin(x))

# Derivada
dy = sp.diff(y, x)

# Integrando comprimento
integrando = sp.sqrt(1+dy**2)

area = sp.simplify(sp.integrate(integrando, (x, pi/6, pi/4)))
area
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
y_lambda = sp.lambdify(x, y, 'numpy')

# x_vals
x_vals = np.linspace(np.sqrt(3)-0.2, np.sqrt(8)+0.2, 1000)

# y_vals
y_vals = y_lambda(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label=r'$y=\ln{x};\:[\sqrt{3},\sqrt{8}]$', color="red")
#ax.plot([])

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(np.sqrt(3), color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(np.sqrt(8), color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
plt.xlim(1.5, 3)
plt.ylim(0.2, 1.2)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$y=\ln{x};\:L=1+\frac{1}{2}\:\ln{\frac{3}{2}}\:u.c.;\:[\sqrt{3},\sqrt{8}]$')
plt.legend()
plt.show()
```
