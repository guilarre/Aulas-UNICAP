```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Ativando o pretty printing do sympy
from sympy import init_printing
init_printing(order='none')
```

# Resoluções

## Questão 1
$\iint_R (x+y)\:dA\:;\:0\le x\le 1,\:0\le y\le 2$

```python
x = sp.Symbol('x')
y = sp.Symbol('y')

f = x + y

integral_dupla = sp.integrate(f, (x, 0, 1), (y, 0, 2))
integral_dupla
```

```python
# Plotando o gráfico

# x_vals, y_vals
x_vals = np.linspace(0, 1, 100)
y_vals = np.linspace(0, 2, 100)

# Criando o gráfico
fig = plt.figure(figsize=(12, 7))

# Subgráfico 1 -> superfície retangular
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
X, Y = np.meshgrid(x_vals, y_vals)
Z = X + Y

# Plotando a superfície
ax1.plot_surface(X, Y, Z, cmap='plasma', alpha=0.8, edgecolor='none')

# Adicionando as bordas da região de integração
ax1.plot([0, 1, 1, 0, 0], [0, 0, 2, 2, 0], [0, 0, 0, 0, 0], color='purple', linewidth=2, label='Região de integração')

ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x,y)')
ax1.set_title('Superfície z = x + y')
ax1.set_xticks([0, 0.5, 1, 1.5, 2], ['0', '0.5', '1', '1.5', '2'])
ax1.set_yticks([0, 0.5, 1, 1.5, 2], ['0', '0.5', '1', '1.5', '2'])
ax1.view_init(10, 45)
ax1.set_xlim(0, 2)
ax1.legend()

# Subgráfico 2 -> VOLUME entre z = f(x,y) (em cima)
# e z = 0 (plano xy em baixo)
ax2 = fig.add_subplot(1, 2, 2, projection='3d')

# Plotando polígonos para representar o volume
vertices = []
# y = 2
for i in range(len(x_vals) - 1):
    v1 = [[x_vals[i], 0, 0], [x_vals[i+1], 0, 0], [x_vals[i+1], 0, Z[0,i+1]], [x_vals[i], 0, Z[0,i]]]
    vertices.append(v1)
# y = 0
for i in range(len(x_vals) - 1):
    v2 = [[x_vals[i], 2, 0], [x_vals[i+1], 2, 0], [x_vals[i+1], 2, Z[-1,i+1]], [x_vals[i], 2, Z[-1,i]]]
    vertices.append(v2)
# x = 1
for j in range(len(y_vals) - 1):
    v3 = [[0, y_vals[j], 0], [0, y_vals[j+1], 0], 
          [0, y_vals[j+1], Z[j+1,0]], [0, y_vals[j], Z[j,0]]]
    vertices.append(v3)
# x = 0
for j in range(len(y_vals) - 1):
    v4 = [[1, y_vals[j], 0], [1, y_vals[j+1], 0], 
          [1, y_vals[j+1], Z[j+1,-1]], [1, y_vals[j], Z[j,-1]]]
    vertices.append(v4)

# Criar coleção de polígonos
poly_collection = Poly3DCollection(vertices, alpha=0.2, color="#ffc59f", label='V = 3 u.v.')
ax2.add_collection3d(poly_collection)

# Base
ax2.plot_surface(X, Y, np.zeros_like(X), alpha=0.2, color='purple')

ax2.set_title('Volume entre superfície e base (z = 0)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')
ax2.plot([0, 1, 1, 0, 0], [0, 0, 2, 2, 0], [0, 0, 0, 0, 0], color='purple', linewidth=2, label='Região de integração')
ax2.plot_surface(X, Y, Z, alpha=0.8, cmap='plasma')
ax2.set_xticks([0, 0.5, 1, 1.5, 2], ['0', '0.5', '1', '1.5', '2'])
ax2.set_yticks([0, 0.5, 1, 1.5, 2], ['0', '0.5', '1', '1.5', '2'])
ax2.view_init(10, 45)
ax2.set_xlim(0, 2)
ax2.legend()

plt.show()
```
