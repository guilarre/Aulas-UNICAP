# Preparação inicial

```python
# Imports

import sympy as sp
from sympy import E, sin, cos, tan, sec, csc, cot
from sympy import init_printing
import numpy as np
import matplotlib.pyplot as plt

# Ativando o pretty printing

init_printing(order='none')

# Definir a variável simbólica x

x = sp.Symbol('x')
```

# Resoluções

## Questão 1

```python
# ∫ k dx

# Definindo uma variável k (constante)

k = sp.Symbol('k')

# Definindo a função

f = k

# Calculando a integral indefinida

integral = sp.integrate(f, x) + sp.Symbol('C')
integral
```

```python
# Criando valores numéricos para plotar
k_val = 2
C_val = 0

# Substituindo valores numéricos na expressão
f_num = f.subs(k, k_val)
integral_num = integral.subs({k: k_val, sp.Symbol('C'): C_val})

# Convertendo expressões simbólicas em numéricas (sp.lambdify())
f_lambdified = sp.lambdify(x, f_num, 'numpy')
integral_lambdified = sp.lambdify(x, integral_num, 'numpy')

# Definindo o intervalo de x para plotar
x_vals = np.linspace(-2, 2, 400)

# Avaliando as funções nos pontos de x_vals
# Para a função constante, criamos um array
# com o mesmo valor para todos os x (.full_like())
f_vals = np.full_like(x_vals, f_lambdified(0))
integral_vals = integral_lambdified(x_vals)

# Plotando a função original e sua primitiva
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label=f'f(x) = $2$', color='blue')
plt.plot(x_vals, integral_vals, label=f'F(x) = $2x + C$', color='red')

# Adicionando rótulos e legenda ao gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Mostrando o gráfico
plt.grid(True)
plt.title('Função e Sua Primitiva')
plt.show()
```

```python
# ∫ sec^2(x) dx

# Definindo a função

f = sec(x)**2

# Calculando a integral indefinida
# Usando o método simplify() para mostrar como tan(x), não sin(x)/cos(x)

integral = sp.integrate(f, x) + sp.Symbol('C')
integral.simplify()
```

```python
# Criando valores numéricos para plotar
C_val = 0

# Funções numéricas
f_func = sp.lambdify(x, f, 'numpy')
integral_func = sp.lambdify(x, integral.subs(sp.Symbol('C'), C_val), 'numpy')

# Definindo múltiplos intervalos para evitar descontinuidades
intervals = [
    (-np.pi/2 + 0.1, np.pi/2 - 0.1),      # Período central
    (np.pi/2 + 0.1, 3*np.pi/2 - 0.1),     # Período à direita
    (-3*np.pi/2 + 0.1, -np.pi/2 - 0.1)    # Período à esquerda
]

# Criando figura com eixos centralizados
fig, ax = plt.subplots(figsize=(14, 6))

# Configurando eixos centralizados
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Adicionando setas
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False, markersize=8)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False, markersize=8)

for i, (x_min, x_max) in enumerate(intervals):
    x_vals = np.linspace(x_min, x_max, 200)
    
    # Avaliando funções
    f_vals = f_func(x_vals)
    F_vals = integral_func(x_vals)
    
    # Plotando
    ax.plot(x_vals, f_vals, color='blue', linewidth=2, 
            label=r'$f(x) = \sec^2(x)$' if i == 0 else "")
    ax.plot(x_vals, F_vals, color='red', linewidth=2, 
            label=r'$F(x) = \tan(x) + C$' if i == 0 else "")

# Adicionando linhas verticais para assíntotas
for k in range(-2, 3):
    asymptote = np.pi/2 + k*np.pi
    ax.axvline(x=asymptote, color='gray', linestyle='--', alpha=0.5, linewidth=1)

# Configurações finais
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.legend(fontsize=12)
ax.grid(True, alpha=0.3)
ax.set_title('Função e sua Primitiva', fontsize=14)
ax.set_xlim(-5, 5)
ax.set_ylim(-10, 10)

# Adicionando ticks em múltiplos de π/2
from matplotlib.ticker import MultipleLocator
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi/2))
ax.set_xticklabels(['-2π', '-3π/2', '-π', '-π/2', '0', 'π/2', 'π', '3π/2', '2π'])

plt.tight_layout()
plt.show()
```

## Questão 9

```python
# ∫ csc^2(x) dx

# Definindo a função

f = csc(x)**2

# Calculando a integral indefinida
# Usando o método simplify() para mostrar como tan(x), não cos(x)/sin(x)
# Usando rewrite() para mostrar como -cot(x), não -1/tan(x) 

integral = sp.integrate(f, x) + sp.Symbol('C')
integral_simplified = integral.simplify().rewrite(cot)
integral_simplified
```

```python
C_val = 0

# Funções numéricas
f_func = sp.lambdify(x, f, 'numpy')
integral_func = sp.lambdify(x, integral_simplified.subs(sp.Symbol('C'), C_val), 'numpy')

# Definindo múltiplos intervalos para evitar descontinuidades (x ≠ kπ)
intervals = [
    (0.1, np.pi - 0.1),           # Período (0, π)
    (np.pi + 0.1, 2*np.pi - 0.1), # Período (π, 2π)
    (-np.pi + 0.1, -0.1)          # Período (-π, 0)
]

# Criando figura com eixos centralizados
fig, ax = plt.subplots(figsize=(14, 6))

# Configurando eixos centralizados
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Adicionando setas
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False, markersize=8)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False, markersize=8)

# Plotando para cada intervalo

for i, (x_min, x_max) in enumerate(intervals):
    x_vals = np.linspace(x_min, x_max, 200)
    
    # Avaliando funções
    f_vals = f_func(x_vals)
    F_vals = integral_func(x_vals)
    
    # Plotando
    ax.plot(x_vals, f_vals, color='blue', linewidth=2, 
            label=r'$f(x) = csc^2(x)$' if i == 0 else "")
    ax.plot(x_vals, F_vals, color='red', linewidth=2, 
            label=r'$F(x) = -cot(x) + C$' if i == 0 else "")

# Adicionando linhas verticais para assíntotas (x = kπ)
for k in range(-2, 3):
    asymptote = k * np.pi
    ax.axvline(x=asymptote, color='gray', linestyle='--', alpha=0.5, linewidth=1)

# Configurações finais
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.legend(fontsize=12, loc='upper right')
ax.grid(True, alpha=0.3)
ax.set_title('Função e sua Primitiva', fontsize=14)
ax.set_xlim(-2*np.pi, 2*np.pi)
ax.set_ylim(-10, 10)

# Adicionando ticks em múltiplos de π
from matplotlib.ticker import MultipleLocator
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi/2))
ax.set_xticklabels(['-2π', '-3π/2', '-π', '-π/2', '0', 'π/2', 'π', '3π/2', '2π'])

plt.tight_layout()
plt.show()
```

```python
# ∫ 1/√(x²-1) dx
f = 1/sp.sqrt(x**2-1)
integral = sp.integrate(f, x) + sp.Symbol('C')
print(f"Integral: {integral}")

# Configurações
C_val = 0
f_func = sp.lambdify(x, f, 'numpy')
integral_func = sp.lambdify(x, integral.subs(sp.Symbol('C'), C_val), 'numpy')

# Domínio: x < -1 ou x > 1
x_vals1 = np.linspace(-5, -1.01, 200)  # Ramo esquerdo
x_vals2 = np.linspace(1.01, 5, 200)    # Ramo direito

f_vals1 = f_func(x_vals1)
f_vals2 = f_func(x_vals2)
integral_vals1 = integral_func(x_vals1)
# cosh^-1 começa em 1, então plotamos essa área
integral_vals2 = integral_func(np.linspace(1, 5, 200))

fig, ax = plt.subplots(figsize=(12, 8))
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False, markersize=8)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False, markersize=8)

# Plotando os dois ramos
ax.plot(x_vals1, f_vals1, color='blue', linewidth=2, label=r'$f(x) = \frac{1}{\sqrt{x^2-1}}$')
ax.plot(x_vals2, f_vals2, color='blue', linewidth=2)
ax.plot(x_vals1, integral_vals1, color='red', linewidth=2, label=r'$F(x) = \cosh^{-1}(x) + C$')
ax.plot(x_vals2, integral_vals2, color='red', linewidth=2)

# Plotando um ponto para melhor visualizar de onde começa F(x)
ax.plot(1, 0, marker='o', color='red', markersize=6)
ax.text(1.1, -0.2, '(1, 0)')

# Assíntotas verticais
ax.axvline(x=-1, color='gray', linestyle='--', alpha=0.7)
ax.axvline(x=1, color='gray', linestyle='--', alpha=0.7)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_title('Função e sua Primitiva')
ax.set_xlim(-5, 5)
ax.set_ylim(-2, 4)
plt.tight_layout()
plt.show()
```
