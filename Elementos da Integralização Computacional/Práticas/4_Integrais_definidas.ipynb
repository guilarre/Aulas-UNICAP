# Preparação inicial

```python
# Imports
import sympy as sp
from sympy import E, sin, cos, tan, pi
from sympy import init_printing
import numpy as np
import matplotlib.pyplot as plt

# Ativando o pretty printing
init_printing(order='none')
```

# Resoluções

## Questão 1

```python
# ∫_0^2 (3x^2) dx

# Declarando a variável x
x = sp.Symbol('x')

# Definindo a função f(x)
f = 3*x**2

# Calculando a integral definida
integral_def = sp.integrate(f, (x, 0, 2))
integral_def
```

```python
# Plotando os gráficos

# Calculando a primitiva F(x)
F = sp.integrate(f, x)

# Convertendo em funções lambda
f_lambdified = sp.lambdify(x, f, 'numpy')
F_lambdified = sp.lambdify(x, F, 'numpy')

# x_vals
x_vals = np.linspace(0, 2.5, 200)

# Calculando as funções para x_vals
f_vals = f_lambdified(x_vals)
F_vals = F_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(12, 5))

# Subgráfico 1: área sob a curva
plt.subplot(1, 2, 1)
plt.plot(x_vals, f_vals, label='$f(x) = 3x^2$', color='orange')
plt.fill_between(x_vals, f_vals, where=(x_vals>=0)&(x_vals<=2), alpha=0.3, color='orange', label=f'A = {integral_def} u.a.')
# Limitando visualmente a área real do gráfico
plt.axhline(0, color="black", linewidth=0.7)
plt.axvline(0, color="black", linewidth=0.7)
plt.title("Integral definida como área")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)

# Subgráfico 2: f(x) e sua integral acumulada
plt.subplot(1, 2, 2)
plt.plot(x_vals, f_vals, label='$f(x) = 3x^2$')
plt.plot(x_vals, F_vals, label='$F(x) = x^3$', linestyle='--')
# Plotando o ponto (2, F(2))
plt.scatter([2], [integral_def], color="red", zorder=3, marker='o', label=f"F(2) = {integral_def}")
# Limitando visualmente a área real do gráfico
plt.axhline(0, color="black", linewidth=0.7)
plt.axvline(0, color="black", linewidth=0.7)
plt.title("Função e integral acumulada")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)

plt.show()
```
