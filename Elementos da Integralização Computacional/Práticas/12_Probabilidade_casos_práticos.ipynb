# Preparações iniciais

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from IPython.display import display, Math

sp.init_printing(order='none')
```

# Resoluções

## Questão 1
$f_T(t) = \frac{1}{100}e^{-t/100},\quad t>0$

(a) Calcule a probabilidade de que uma lâmpada dure mais de 200 horas, P (T > 200).

(b) Calcule a esperança E[T].

```python
# Letra a)

t = sp.Symbol('t')
fa = sp.Rational(1, 100)*sp.exp((-t/100))
P = sp.integrate(fa, (t, 200, sp.oo))
P
```

```python
# Printando em latex

latex_str = "e^{-2} = 0.1353"
display(Math(latex_str))
```

```python
# Letra b)

fb = fa * t
E = sp.integrate(fb, (t, 0, sp.oo))
E
```

```python
# Gráfico da a)
# Valores de t e y
fa_lambda = sp.lambdify(t, fa, 'numpy')
t_vals = np.linspace(0, 500, 500)
y_vals = fa_lambda(t_vals)
# Subgráfico 1: curva
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(t_vals, y_vals, label=r'$f_T(t) = \frac{1}{100}e^{-t/100};\:[0,\infty]$')
# Retas/assíntotas
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7)
# Opções
plt.xlim(0, 500)
plt.grid(linestyle='--', alpha=0.3)
plt.title(r"Curva da densidade do tempo de vida de lâmpadas LED (x) entre 0 e $\infty$ horas")
plt.legend()

# Gráfico da b)
# Subgráfico 2: área sob curva P(T > 200) e esperança E(T)
plt.subplot(2, 1, 2)
plt.plot(t_vals, y_vals, label=r'$f_T(t) = \frac{1}{100}e^{-t/100};\:[0,\infty]$')
plt.fill_between(t_vals, y_vals, where=(t_vals>=200)&(t_vals<=500), label=r'$P(T\:>\:200) = e^{-2} = 0,1353$', color='lightblue')
# E(T)
plt.axvline(100, label='E(T) = 100h', color='green')
# Retas/assíntotas
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7)
plt.axvline(200, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Opções
plt.xlim(0, 500)
plt.grid(linestyle='--', alpha=0.3)
plt.title("Probabilidade que lâmpada dure mais que 200h e esperança E(T)")
plt.legend()

# Chamando plotagem
plt.tight_layout()
plt.show()
```

## Questão 2
$f_W(w) = 3w^2,\quad 0\leq w\leq 1$

(a) Verifique que fW é uma densidade válida determinando a constante (se necessário).

(b) Calcule P (W > 0.8) e E[W].

```python
# Letra a)

w = sp.Symbol('w')
fa = 3*w**2
P = sp.integrate(fa, (w, 0, 1))
P
```

```python
# Letra b), parte 1

P = sp.integrate(fa, (w, 0.8, 1))
P
```

```python
# Letra b), parte 2

fb = fa * w
E = sp.integrate(fb, (w, 0, 1))
E
```

```python
latex_str = r"\frac{3}{4}s = 45s"
display(Math(latex_str))
```

```python
# Gráfico da a)
# Valores de w e y
fa_lambda = sp.lambdify(w, fa, 'numpy')
w_vals = np.linspace(0, 1, 500)
y_vals = fa_lambda(w_vals)
# Subgráfico 1: curva
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(w_vals, y_vals, label=r'$f_w(w) = 3w^2;\:[0,1]$')
# Retas/assíntotas
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7)
# Opções
plt.xlim(0, 1)
plt.grid(linestyle='--', alpha=0.3)
plt.title("Densidade do tempo de espera em caixa eletrônico (w) entre 0 e 1 minutos")
plt.legend()

# Gráfico da b)
# Subgráfico 2: área sob curva P(W > 200) e esperança E(W)
plt.subplot(2, 1, 2)
plt.plot(w_vals, y_vals, label=r'$f_W(w) = 3w^2;\:[0,1]$')
plt.fill_between(w_vals, y_vals, where=(w_vals>=0.8)&(w_vals<=1), label=r'$P(0.8\leq w\leq 1) = 0,488$', color='lightblue')
# E(W)
plt.axvline((3/4), label=r'E(W) = $\frac{3}{4}$m = 45s', color='green')
# Retas/assíntotas
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7)
plt.axvline(0.8, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Opções
plt.xlim(0, 1)
plt.grid(linestyle='--', alpha=0.3)
plt.title("P do tempo de espera (w) 0.8 e 1 minutos e esperança E(W)")
plt.legend()

# Chamando plotagem
plt.tight_layout()
plt.show()
```

## Questão 3
$f_R(r) = \frac{1}{2\sqrt{r}},\quad 0\lt r\lt 1$

(a) Mostre que é uma função densidade.

(b) Calcule P (0.25 < R < 0.81) e E[R].

```python
# Letra a)

r = sp.Symbol('r')
fa = sp.Rational(1, 2)*1/sp.sqrt(r)
P = sp.integrate(fa, (r, 0, 1))
P
```

```python
# Letra b), parte 1

P = sp.integrate(fa, (r, 0.25, 0.81))
P
```

```python
# Letra b), parte 2

fb = fa * r
E = sp.integrate(fb, (r, 0, 1))
E
```

```python
latex_str = r"\frac{1}{3}\:horas = 20\:minutos"
display(Math(latex_str))
```

```python
# Gráfico da a)
# Valores de r e y
fa_lambda = sp.lambdify(r, fa, 'numpy')
r_vals = np.linspace(0, 1, 100)
y_vals = fa_lambda(r_vals)
# Subgráfico 1: curva
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(r_vals, y_vals, label=r'$f_r(r) = \frac{1}{2\sqrt{r}};\:[0,1]$')
# Retas/assíntotas
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7)
# Opções
plt.xlim(0, 1)
plt.grid(linestyle='--', alpha=0.3)
plt.title("Densidade do tempo de recarga de uma bateria (r) entre 0 e 1 hora")
plt.legend()

# Gráfico da b)
# Subgráfico 2: área sob curva P(R > 200) e esperança E(R)
plt.subplot(2, 1, 2)
plt.plot(r_vals, y_vals, label=r'$f_R(r) = \frac{1}{2\sqrt{r}};\:[0,1]$')
plt.fill_between(r_vals, y_vals, where=(r_vals>=0.25)&(r_vals<=0.81), label=r'$P(0.25\leq w\leq 0.81) = 0,4$', color='lightblue')
# E(W)
plt.axvline((1/3), label=r'E(R) = $\frac{1}{3}$h = 20m', color='green')
# Retas/assíntotas
plt.axhline(0, color='black', alpha=0.5, linewidth=0.7)
plt.axvline(200, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Opções
plt.xlim(0, 1)
plt.grid(linestyle='--', alpha=0.3)
plt.title("P do tempo de espera (w) 0 e 1 minutos e esperança E(W)")
plt.legend()

# Chamando plotagem
plt.tight_layout()
plt.show()
```
