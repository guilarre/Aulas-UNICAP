## Questão 1

```python
# Questão 1

# Definindo a variável simbólica
x = symbols('x')

# Definindo a função
func = 5*x**3 - 2*x**2 + 7*x - 4

# Calculando a derivada da função
derivada = sp.diff(func, x)

# Mostrando a função derivada na saída
derivada
```

```python
# Convertendo em funções lambda para podermos plotar
func_lambdified = sp.lambdify(x, func, 'numpy')
func_prime_lambdified = sp.lambdify(x, derivada, 'numpy')

# Intervalo de x para plotagem
x_vals = np.linspace(-3, 3, 400)

# Calculando os valores da função e de sua derivada nos pontos x_vals
f_vals = func_lambdified(x_vals)
f_prime_vals = func_prime_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Gráfico da função original
plt.plot(x_vals, f_vals, label='Função: $f(x) = 5x^3 - 2x^2 + 7x - 4$')

# Gráfico da derivada
plt.plot(x_vals, f_prime_vals, label='Derivada: $f\'(x) = 15x^2 - 4x + 7$', linestyle='dashed')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Função e sua Derivada')
plt.legend()
plt.grid(True)
plt.show()
```

## Questão 11

```python
# Questão 11

# Definindo a variável simbólica
t = symbols('t')

# Definindo a função
movimento = t**3 - 6*(t**2) + 9*t

# Calculando a derivada da função para obtermos a função velocidade
velocidade = sp.diff(movimento, t)

# Mostrando a função derivada na saída
velocidade

# Convertendo a função velocidade em uma função lambda para podermos substituir
# o valor de 't' com eficiência: sp.lambdify(variables, expression, library)
velocidade_lambdified = sp.lambdify(t, velocidade, 'numpy')

# Outra opção sem usar a conversão em função lambda
# é usar o método .subs(symbol, value) de um objeto Sympy
# (menos eficiente para operações mais complexas, como plotagem, mas
# funciona bem quando há apenas uma variável):
# velocidade.subs(t, 2)

# Determinando a velocidade no instante t = 2s
velocidade_lambdified(2)
```

```python
# Convertendo em função lambda para podermos plotar
movimento_lambdified = sp.lambdify(t, movimento, 'numpy')

# Intervalo de x para plotagem
t_vals = np.linspace(0, 4, 400)

# Calculando os valores da função e de sua derivada nos pontos t_vals
movimento_vals = movimento_lambdified(t_vals)
velocidade_vals = velocidade_lambdified(t_vals)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Plotando o ponto t = 2s
x0, y0 = 2, velocidade_lambdified(2)
plt.plot(x0, y0, 'ro')
plt.text(x0+0.1, y0+0.5, f"({x0}, {y0})")

# Gráfico da função original
plt.plot(t_vals, movimento_vals, label=r'Movimento: $s(t) = t^3 - 6t^2 + 9t$')

# Gráfico da derivada
plt.plot(t_vals, velocidade_vals, label=r'Velocidade: $v(t) = 3t^2 - 12t + 9$', linestyle='dashed')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Funções movimento e velocidade')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculando o valor do coeficiente angular para o ponto x = 1
coef_angular_lambdified = sp.lambdify(x, coef_angular, 'numpy')
coef_angular_lambdified(1)
```

```python
# Convertendo em função lambda para podermos plotar
func_lambdified = sp.lambdify(x, func, 'numpy')

# Intervalo de x para plotagem
x_vals = np.linspace(-3, 3, 400)

# Calculando os valores da função e de sua derivada nos pontos x_vals
func_vals = func_lambdified(x_vals)
coef_angular_vals = coef_angular_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Plotando o ponto x = 1
x0, y0 = 1, coef_angular_lambdified(1)
plt.plot(x0, y0, 'ro')
plt.text(x0-0.1, y0+0.1, f"({x0}, {int(y0)})")

# Gráfico da função original
plt.plot(x_vals, func_vals, label=r'Função: $y = ln(x^2 + 1)$')

# Gráfico da derivada
plt.plot(x_vals, coef_angular_vals, label=r'Derivada: $y\' = \frac{2x}{x^2+1}$', linestyle='dashed')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Função e sua derivada')
plt.legend()
plt.grid(True)
plt.show()
```
