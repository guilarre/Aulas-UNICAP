# Preparação inicial

```python
# Imports
import sympy as sp
from sympy import sin, cos, pi
from sympy import init_printing
import numpy as np
import matplotlib.pyplot as plt

# Ativando o pretty printing
init_printing(order='none')
```

# Resoluções

## Áreas entre curvas

## Questão 1
$y_1 = 2-x^2\:;\:y_2 = -x\:;\:[-1,2]$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')

# Definindo as funções
y1 = 2 - x**2
y2 = -x

# Área desejada
area = sp.integrate((y1-y2), (x, -1, 2))
area
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
y1_lambda = sp.lambdify(x, y1, 'numpy')
y2_lambda = sp.lambdify(x, y2, 'numpy')

# x_vals
x_vals = np.linspace(-1.3, 2.3, 1000)

# y1_vals, y2_vals
y1_vals = y1_lambda(x_vals)
y2_vals = y2_lambda(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y1_vals, label=r'$y_1 = 2-x^2$', color="red")
plt.plot(x_vals, y2_vals, label=r'$y_2 = -x$', color="orange")

# Área hachurada
plt.fill_between(x_vals, y1_vals, y2_vals, where=(x_vals>=-1)&(x_vals<=2), color='orange', alpha=0.4, label=r'A = $\frac{9}{2}$u.a.')

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(-1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(2, color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
plt.xlim(-1.3, 2.3)
plt.ylim(-2.3, 2.5)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'Área entre $y_1$ e $y_2$; $[-1,2]$')
plt.legend()
plt.show()
```

## Questão 2
$y=0\:;\:x_1=y^2\:;\:x_2=y+2\:;\:[0,2]$

```python
# Definindo as variáveis simbólicas
y = sp.Symbol('y')

# Definindo as funções
x1 = y**2
x2 = y + 2

# Área desejada
area = sp.integrate((x2-x1), (y, 0, 2))
area
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
x1_lambda = sp.lambdify(y, x1, 'numpy')
x2_lambda = sp.lambdify(y, x2, 'numpy')

# y_vals
y_vals = np.linspace(0, 2, 1000)

# x1_vals, x2_vals
x1_vals = x1_lambda(y_vals)
x2_vals = x2_lambda(y_vals)

# Criando o gráfico
plt.figure(figsize=(12, 8))
plt.plot(x1_vals, y_vals, label=r'$x_1 = y^2$', color="red")
plt.plot(x2_vals, y_vals, label=r'$x_2 = y+2$', color="orange")
plt.hlines(0, 0, 2, color='blue', label=r'$y=0;\:[0,2]$', zorder=3)

# Área hachurada
plt.fill_betweenx(y_vals, x1_vals, x2_vals, where=(y_vals>=0)&(y_vals<=2), color='orange', alpha=0.4, label=r'A = $\frac{10}{3}$u.a.')

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(4, color='black', alpha=0.5, linewidth=0.7, linestyle='--')

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'Área entre $x_1$ e $x_2$; $[0,2]$')
plt.legend()
plt.show()
```

## Questão 3
$y_1=x\sqrt{4-x^2}\:;\:y_2=0\:;\:[-2,2]$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')

# Definindo as funções
y = x*sp.sqrt(4 - x**2)

# Área desejada
area1 = sp.Abs(sp.integrate((y), (x, -2, 0))) # Módulo
area2 = sp.integrate((y), (x, 0, 2))
area = area1 + area2
area
```

```python
# Plotando o gráfico

# Criando função lambda para plotar
y_lambda = sp.lambdify(x, y, 'numpy')

# x_vals
x_vals = np.linspace(-2, 2, 1000)

# y_vals
y_vals = y_lambda(x_vals)

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label=r'$y_1=x\sqrt{4-x^2}$', color="red")
plt.axhline(0, color="orange", label=r'$y_2=0$', zorder=3)

# Área hachurada
plt.fill_between(x_vals, y_vals, where=(x_vals>=-2)&(x_vals<=2), color='orange', alpha=0.4, label=r'A = $\frac{16}{3}$u.a.')

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(-2, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(2, color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'Área entre $y_1$ e $y_2$; $[-2,2]$')
plt.legend()
plt.show()
```

## Questão 9
$y_1=4-x^2\:;\:y_2=-x+2\:;\:x_1=-2\:;\:x_2=3\:;\:[-2,3]$

```python
# Definindo as variáveis simbólicas
x = sp.Symbol('x')

# Definindo as funções
y1 = 4 - x**2
y2 = -x + 2

# Área desejada
area1 = sp.integrate((y2-y1), (x, -2, -1))
area2 = sp.integrate((y1-y2), (x, -1, 2))
area3 = sp.integrate((y2-y1), (x, 2, 3))
area = area1 + area2 + area3
area
```

```python
# Plotando o gráfico

# Criando funções lambda para plotar
y1_lambda = sp.lambdify(x, y1, 'numpy')
y2_lambda = sp.lambdify(x, y2, 'numpy')

# x_vals
x1_vals = np.linspace(-3, 3, 1000)
x2_vals = np.linspace(-3, 5, 1000)

# y1_vals, y2_vals
y1_vals = y1_lambda(x1_vals)
y2_vals = y2_lambda(x2_vals)

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x1_vals, y1_vals, label=r'$y_1=4-x^2$', color="red")
plt.plot(x2_vals, y2_vals, label=r'$y_2=-x+2$', color="orange")
plt.vlines(-2, 0, 4, color='blue', label=r'$x_1=-2;\:[0,4]$')
plt.vlines(3, -1, -5, color='blue', label=r'$x_2=3;\:[-1,5]$')

# Área hachurada
x_vals = x2_vals
y1_vals = y1_lambda(x_vals)
y2_vals = y2_lambda(x_vals)
plt.fill_between(x_vals, y2_vals, y1_vals, where=((x_vals>=-2)&(x_vals<=-1))|((x_vals>=2)&(x_vals<=3)), color='orange', alpha=0.4, label=r'A = $\frac{49}{6}$u.a.')
plt.fill_between(x_vals, y1_vals, y2_vals, where=(x_vals>=-1)&(x_vals<=2), color='orange', alpha=0.4)

# Eixos x, y
plt.axvline(0, color='black', alpha=0.5, linewidth=1)
plt.axhline(0, color='black', alpha=0.5, linewidth=1)
plt.axvline(-2, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(3, color='black', alpha=0.5, linewidth=0.7, linestyle='--')

# Diminuindo a escala
plt.xlim(-3, 5)
plt.ylim(-5, 5)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'Área entre $y_1$ e $y_2$; $[-2,3]$')
plt.legend()
plt.show()
```
