# Preparações iniciais

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Ativando o pretty printing do sympy
from sympy import init_printing
init_printing(order='none')
```

# Resoluções

## Questão 1
$f(x) = \begin{cases}
    \frac{1}{4},\:0\le x\le 4\\
    0,\:\text{caso contrário.}
\end{cases}$

(a) Verifique se f(x) é uma função densidade válida.

(b) Calcule P (1 ≤ X ≤ 3) utilizando integrais.

(c) Represente graficamente a densidade e a área correspondente à probabilidade
calculada.

```python
# Letra a)

x = sp.Symbol('x')
f = sp.Rational(1, 4)
P = sp.integrate(f, (x, 0, 4))
P
```

```python
# Letra b)

P = sp.integrate(f, (x, 1, 3))
P
```

```python
# Letra c)

f_lambda = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(0, 4, 100)
y_vals = (1/4) * np.ones_like(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{1}{4}$', color='orange')

x_vals = np.linspace(1, 3, 100)
y_vals = (1/4) * np.ones_like(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(1 \leq x \leq 3) = \frac{1}{2} = 50\%$', color='orange', alpha=0.4)

plt.vlines(1, 0, 1, color='black', linewidth=0.7, linestyle='--')
plt.vlines(3, 0, 1, color='black', linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', linewidth=0.7, alpha=0.5)
plt.axhline(0, color='black', linewidth=0.7, alpha=0.5)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$P(1 \leq x \leq 3)$')
plt.legend()
plt.show()
```

## Questão 2
$f(x) = \lambda e^{-\lambda x},\:x \ge 0,\:\lambda = 0.5$

(a) Calcule a probabilidade P (X ≤ 3).

(b) Determine P (2 ≤ X ≤ 5) por meio de integrais.

(c) Esboce o gráfico da função densidade e destaque a região correspondente.

```python
# Letra a)

x = sp.Symbol('x')
f = (sp.Rational(1, 2))*sp.exp(-sp.Rational(1, 2)*x)
P = sp.integrate(f, (x, 0, 3))
P
```

```python
# Letra b)

P = sp.integrate(f, (x, 2, 5))
P
```

```python
# Letra c)

f_lambda = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(-1, 6, 200)
y_vals = f_lambda(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{1}{2}e^{\frac{-x}{2}}$', color='orange')

x_vals = np.linspace(0, 3, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(X \leq 3) = 1-e^{\frac{-3}{2}}$', color='orange', alpha=0.4)

x_vals = np.linspace(2, 5, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(2 \leq X \leq 5) = e^{-1}-e^{\frac{-5}{2}}$', color='red', alpha=0.2)

# plt.vlines(2, 0, 1, color='black', linewidth=0.7, linestyle='--')
# plt.vlines(3, 0, 1, color='black', linewidth=0.7, linestyle='--')
# plt.vlines(5, 0, 1, color='black', linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', linewidth=0.7, alpha=0.5)
plt.axhline(0, color='black', linewidth=0.7, alpha=0.5)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$P(1 \leq x \leq 3)$')
plt.legend()
plt.show()
```

## Questão 3
$f(x) = \begin{cases}
    kx^2,\:0 \le x \le 2\\
    0,\:\text{caso contrário.}
\end{cases}$

(a) Determine o valor da constante k para que f(x) seja uma densidade válida.

(b) Calcule P (1 ≤ X ≤ 2).

(c) Interprete geometricamente o resultado obtido.

```python
# Letra a)

x = sp.Symbol('x')
k = sp.Symbol('k')
f = k*x**2
integral = sp.solve(sp.Eq(sp.integrate(f, (x, 0, 2)), 1))[0]
integral
```

```python
# Letra b)

P = sp.integrate(f.subs({k: integral}), (x, 1, 2))
P
```

```python
# Letra c)

f = f.subs({k: integral})
f_lambda = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(0, 2.5, 200)
y_vals = f_lambda(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{3x^2}{8}$', color='orange')

x_vals = np.linspace(1, 2, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(1 \leq X \leq 2) = \frac{7}{8}$', color='orange', alpha=0.4)

plt.vlines(1, 0, 3, color='black', linewidth=0.7, linestyle='--')
plt.vlines(2, 0, 3, color='black', linewidth=0.7, linestyle='--')
# plt.axvline(0, color='black', linewidth=0.7, alpha=0.5)
# plt.axhline(0, color='black', linewidth=0.7, alpha=0.5)

plt.xlim(0, 2.5)
plt.ylim(0, 3)
plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$P(1 \leq x \leq 2)$')
plt.legend()
plt.show()
```

```python
# Letra b) parte 2

P = sp.integrate(f, (x, 1, 3))
P
```

```python
# Letra c)

f_lambda = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(-1, 5, 200)
y_vals = f_lambda(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{1}{8}(4-x)$', color='orange')

x_vals = np.linspace(0, 2, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(X \leq 2) = \frac{3}{4}$', color='orange', alpha=0.4)

x_vals = np.linspace(1, 3, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(1 \leq X \leq 3) = \frac{1}{2}$', color='red', alpha=0.2)

# plt.vlines(1, 0, 3, color='black', linewidth=0.7, linestyle='--')
# plt.vlines(2, 0, 3, color='black', linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', linewidth=0.7, alpha=0.5)
plt.axhline(0, color='black', linewidth=0.7, alpha=0.5)

plt.xlim(-1, 5)
plt.ylim(-0.2, 1)
plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$P(1 \leq x \leq 2)$')
plt.legend()
plt.show()
```

## Questão 5
$f(x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}$

(a) Escreva a expressão da função densidade

(b) Calcule, utilizando a função de distribuição acumulada, as probabilidades:
- P (−1 ≤ X ≤ 1),
- P (X ≥ 2)

(c) Explique o significado geométrico dessas integrais no contexto da área sob a
curva.

```python
# Letra b) parte 1

# O sympy usa o sp.erf() (função de erro Gaussiana) para calcular a FDA da normal padrão (phi)
def phi(x_val):
    return (1 + sp.erf(x_val/sp.sqrt(2)))/2

P = (phi(1)-phi(-1)).evalf()
P
```

```python
# Letra b) parte 2

P = 1-phi(2).evalf()
P
```

```python
# Escrevendo a função da distribuição normal padrão

f = 1/sp.sqrt(2*sp.pi) * sp.exp(-x**2/2)
f
```

```python
# Letra c)

f_lambda = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(-3, 3, 200)
y_vals = f_lambda(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}$', color='orange')

x_vals = np.linspace(-1, 1, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(-1 \leq X \leq 1) = 0,6826$', color='orange', alpha=0.4)

x_vals = np.linspace(2, 3, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(X \geq 2) = 0,0228$', color='red', alpha=0.2)

# plt.vlines(1, 0, 3, color='black', linewidth=0.7, linestyle='--')
# plt.vlines(2, 0, 3, color='black', linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', linewidth=0.7, alpha=0.5)
plt.axhline(0, color='black', linewidth=0.7, alpha=0.5)

plt.xlim(-3, 3)
# plt.ylim(-0.2, 1)
plt.grid(linestyle='--', alpha=0.3)
plt.title(r'FDA da distribuição normal padrão')
plt.legend()
plt.show()
```

## Questão 6
$f(x) = \begin{cases}
    kx(2-x),\:0 \le x \le 2\\
    0,\:\text{caso contrário.}
\end{cases}$

(a) Determine o valor de k.

(b) Calcule P (0.5 ≤ X ≤ 1.5).

(c) Esboce o gráfico da densidade e destaque a área correspondente.

```python
# Letra a)

x = sp.Symbol('x')
k = sp.Symbol('k')
f = k*x*(2-x)
integral = sp.solve(sp.Eq(sp.integrate(f, (x, 0, 2)), 1))[0]
integral
```

```python
# Letra b)

P = sp.integrate(f.subs({k: sp.Rational(3, 4)}), (x, sp.Rational(1, 2), sp.Rational(3, 2)))
P
```

```python
# Letra c)

f_lambda = sp.lambdify(x, f.subs({k: sp.Rational(3, 4)}), 'numpy')
x_vals = np.linspace(-1, 3, 200)
y_vals = f_lambda(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{3}{4}x(2-x)$', color='orange')

x_vals = np.linspace(0.5, 1.5, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(0,5 \leq X \leq 1,5) = \frac{11}{16}$', color='orange', alpha=0.4)

# plt.vlines(1, 0, 3, color='black', linewidth=0.7, linestyle='--')
# plt.vlines(2, 0, 3, color='black', linewidth=0.7, linestyle='--')
plt.axvline(0.5, color='black', linewidth=0.7, alpha=0.5)
plt.axvline(1.5, color='black', linewidth=0.7, alpha=0.5)
plt.axvline(0, color='black', linewidth=0.7, alpha=0.5)
plt.axhline(0, color='black', linewidth=0.7, alpha=0.5)

# plt.xlim(-1, 5)
# plt.ylim(-0.2, 1)
plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$P(0.5 \leq x \leq 1.5)$')
plt.legend()
plt.show()
```
