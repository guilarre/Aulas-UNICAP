# Preparações iniciais

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Ativando o pretty printing do sympy
from sympy import init_printing
init_printing(order='none')
```

# Resoluções

## Questão 1
$f(x) = \begin{cases}
    \frac{1}{4},\:0\le x\le 4\\
    0,\:\text{caso contrário.}
\end{cases}$

(a) Verifique se f(x) é uma função densidade válida.

(b) Calcule P (1 ≤ X ≤ 3) utilizando integrais.

(c) Represente graficamente a densidade e a área correspondente à probabilidade
calculada.

```python
# Letra a)
x = sp.Symbol('x')
f = 1/4
P = sp.integrate(f, (x, 0, 4))
P
```

```python
# Letra b)
P = sp.integrate(f, (x, 1, 3))
P
```

```python
# Letra c)

f_lambda = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(0, 4, 100)
y_vals = (1/4) * np.ones_like(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{1}{4}$', color='orange')

x_vals = np.linspace(1, 3, 100)
y_vals = (1/4) * np.ones_like(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(1 \leq x \leq 3) = \frac{1}{2} = 50\%$', color='orange', alpha=0.4)

plt.vlines(1, 0, 1, color='black', linewidth=0.7, linestyle='--')
plt.vlines(3, 0, 1, color='black', linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', linewidth=0.7, alpha=0.5)
plt.axhline(0, color='black', linewidth=0.7, alpha=0.5)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$P(1 \leq x \leq 3)$')
plt.legend()
plt.show()
```

## Questão 2
$f(x) = \lambda e^{-\lambda x},\:x \ge 0,\:\lambda = 0.5$

(a) Calcule a probabilidade P (X ≤ 3).

(b) Determine P (2 ≤ X ≤ 5) por meio de integrais.

(c) Esboce o gráfico da função densidade e destaque a região correspondente.

```python
# Letra a)
x = sp.Symbol('x')
f = (sp.Rational(1, 2))*sp.exp(-sp.Rational(1, 2)*x)
P = sp.integrate(f, (x, 0, 3))
P
```

```python
# Letra b)
P = sp.integrate(f, (x, 2, 5))
P
```

```python
# Letra c)

f_lambda = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(-1, 6, 200)
y_vals = f_lambda(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{1}{2}e^{\frac{-x}{2}}$', color='orange')

x_vals = np.linspace(0, 3, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(X \leq 3) = 1-e^{\frac{-3}{2}}$', color='orange', alpha=0.4)

x_vals = np.linspace(2, 5, 200)
y_vals = f_lambda(x_vals)
plt.fill_between(x_vals, y_vals, 0, label=r'$P(2 \leq X \leq 5) = e^{-1}-e^{\frac{-5}{2}}$', color='red', alpha=0.2)

# plt.vlines(1, 0, 1, color='black', linewidth=0.7, linestyle='--')
# plt.vlines(3, 0, 1, color='black', linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', linewidth=0.7, alpha=0.5)
plt.axhline(0, color='black', linewidth=0.7, alpha=0.5)

plt.grid(linestyle='--', alpha=0.3)
plt.title(r'$P(1 \leq x \leq 3)$')
plt.legend()
plt.show()
```
