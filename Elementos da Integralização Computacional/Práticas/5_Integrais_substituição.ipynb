# Preparação inicial

```python
# Imports
import sympy as sp
from sympy import E, sin, cos, tan, sec, pi, ln
from sympy import init_printing
import numpy as np
import matplotlib.pyplot as plt

# Ativando o pretty printing
init_printing(order='none')
```

# Resoluções

## Questão 1
$\int_0^1 2x(x^2+1)^3dx$

```python
# Declarando a variável x
x = sp.Symbol('x')

# Definindo a função f(x)
f = 2*x*(x**2 + 1)**3
# Calculando a integral definida
integral_def = sp.integrate(f, (x, 0, 1))
integral_def
```

```python
# Calculando a primitiva F(x)
F = sp.integrate(f, x)
F
```

```python
# Plotando os gráficos

# Convertendo em funções lambda
f_lambdified = sp.lambdify(x, f, 'numpy')
F_lambdified = sp.lambdify(x, F, 'numpy')

# x_vals
x_vals = np.linspace(-0.5, 1.5, 200)

# Calculando as funções para x_vals
f_vals = f_lambdified(x_vals)
F_vals = F_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(12, 5))

# Subgráfico 1: área sob a curva
plt.subplot(1, 2, 1)
plt.plot(x_vals, f_vals, label=r'$f(x) = 2x(x^2+1)^3$', color='orange')
plt.fill_between(x_vals, f_vals, where=(x_vals>=0)&(x_vals<=1), alpha=0.3, color='orange', label=r'A = $\frac{15}{4}$ u.a.')
# Limitando visualmente a área real do gráfico
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.title("Integral definida como área")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)

# Subgráfico 2: f(x) e sua integral acumulada
# Definir a primitiva acumulada (0 -> x)
F_acumulada = sp.integrate(f, (x, 0, x))
F_acumulada_lamb = sp.lambdify(x, F_acumulada, 'numpy')
F_acumulada_vals = F_acumulada_lamb(x_vals)
plt.subplot(1, 2, 2)
plt.plot(x_vals, f_vals, label=r'$f(x) = 2x(x^2+1)^3$')
plt.plot(x_vals, F_acumulada_vals, label=r'$F(x) = \frac{x^8}{4}+x^6+\frac{3x^4}{2}+x^2$', linestyle='--')
# Plotando o ponto (1, F(1))
plt.scatter([1], [F_acumulada_lamb(1)], color="red", zorder=3, marker='o', label=r"$F(1) = \frac{15}{4}$")
# Limitando visualmente a área real do gráfico
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.title("Função e integral acumulada")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)

plt.show()
```

## Questão 2
$\int_0^\frac{\pi}{4} \tan{x} \sec^2{x}dx$

```python
# Declarando a variável x
x = sp.Symbol('x')

# Definindo a função f(x)
f = tan(x)*sec(x)**2
# Calculando a integral definida
integral_def = sp.integrate(f, (x, 0, pi/4))
integral_def
```

```python
# Calculando a primitiva F(x)
F = (sp.integrate(f, x) - sp.Rational(1, 2)).simplify()
F
```

```python
# Plotando os gráficos

# Convertendo em funções lambda
f_lambdified = sp.lambdify(x, f, 'numpy')
F_lambdified = sp.lambdify(x, F, 'numpy')

# x_vals
x_vals = np.linspace(0, np.pi/4, 200)

# Calculando as funções para x_vals
f_vals = f_lambdified(x_vals)
F_vals = F_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(12, 5))

# Subgráfico 1: área sob a curva
plt.subplot(1, 2, 1)
plt.plot(x_vals, f_vals, label=r'$f(x) = \tan{x}\sec^2{x}$', color='orange')
plt.fill_between(x_vals, f_vals, where=(x_vals>=0)&(x_vals<=np.pi/4), alpha=0.3, color='orange', label=r'A = $\frac{1}{2}$ u.a.')
# Limitando visualmente a área real do gráfico
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(np.pi/4, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Adicionando os ticks relevantes
plt.xticks([0, np.pi/4], ['0', r'$\frac{\pi}{4}$'])
plt.title("Integral definida como área")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)

# Subgráfico 2: f(x) e sua integral acumulada
# Definir a primitiva acumulada (0 -> x)
F_acumulada = sp.integrate(f, (x, 0, x))
F_acumulada_lamb = sp.lambdify(x, F_acumulada, 'numpy')
F_acumulada_vals = F_acumulada_lamb(x_vals)
plt.subplot(1, 2, 2)
plt.plot(x_vals, f_vals, label=r'$f(x) = \tan{x}\sec^2{x}$')
plt.plot(x_vals, F_acumulada_vals, label=r'$F(x) = \frac{\tan^2{x}}{2}$', linestyle='--')
# Plotando o ponto (pi/4, F(pi/4))
plt.scatter([np.pi/4], [F_acumulada_lamb(np.pi/4)], color="red", zorder=3, marker='o', label=r"$F(\frac{\pi}{4}) = \frac{1}{2}$")
# Limitando visualmente a área real do gráfico
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(np.pi/4, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
# Adicionando os ticks relevantes
plt.xticks([0, np.pi/4], ['0', r'$\frac{\pi}{4}$'])
plt.title("Função e integral acumulada")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)

plt.show()

# Gráfico da função em um ciclo (0 -> 2pi), para y > 0
x_vals_1 = np.linspace(0, np.pi/2, 200, endpoint=False)
x_vals_2 = np.linspace(np.pi/2+0.01, np.pi, 200, endpoint=False)
x_vals_3 = np.linspace(np.pi+0.01, (3*np.pi)/2, 200, endpoint=False)
x_vals_4 = np.linspace((3*np.pi)/2+0.01, 2*np.pi, 200, endpoint=False)
plt.plot(x_vals_1, f_vals, label=r'$f(x) = \tan{x}\sec^2{x}$', color='orange')
plt.plot(x_vals_2, f_vals, color='orange')
plt.plot(x_vals_3, f_vals, color='orange')
plt.plot(x_vals_4, f_vals, color='orange')
plt.fill_between(x_vals_1, f_vals, where=(x_vals_1>=0)&(x_vals_1<=np.pi/4), alpha=0.3, color='orange', label=r'A = $\frac{1}{2}$ u.a.')
# Limitando visualmente a área real do gráfico
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(np.pi/4, color='black', alpha=0.9, linewidth=0.7, linestyle='--')
# Adicionando os ticks relevantes
plt.xticks([0, np.pi/4, np.pi/2, np.pi, (3*np.pi)/2, 2*np.pi], ['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
plt.title(r"$f(x) = \tan{x}\sec^2{x}$; [0, $2\pi]$")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)
plt.show()
```

## Questão 10
$\int_1^e \frac{1}{x}e^{\ln{x}}dx$

```python
# Declarando a variável x
x = sp.Symbol('x')

# Definindo a função f(x)
f = sp.exp(sp.ln(x))/x

# Calculando a integral definida
integral_def = sp.integrate(f, (x, 1, E))
integral_def
```

```python
# Calculando a primitiva F(x)
F = sp.integrate(f, x)
F
```

```python
# Plotando os gráficos

# Convertendo em funções lambda
f_lambdified = sp.lambdify(x, f, 'numpy')
F_lambdified = sp.lambdify(x, F, 'numpy')

# x_vals
x_vals = np.linspace(1, np.e, 200)


# Calculando as funções para x_vals

# No caso de f_vals, é necessário criar uma função numérica manualmente, sem o Sympy.
# Usar o sp.lambdify deu erro com essa função (pq o Sympy retorna apenas um 1,
# portanto só tem 1 dimensão, VS. 200 dimensões do x_vals). Da forma abaixo,
# retornamos o 1 200 vezes para podermos traçar uma linha reta no gráfico.

# De acordo com o Deepseek: "O lambdify é ótimo para expressões complexas, mas pode ter
# comportamentos inesperados com simplificações automáticas. Para expressões mais simples
# ou quando você quer preservar a forma original, definir a função manualmente é mais seguro
# e previsível. A função direta também é mais eficiente computacionalmente, pois evita o
# overhead da conversão simbólica!"

# Definindo manualmente a função
def f_numeric(x):
    return (1/x) * np.exp(np.log(x))

f_vals = f_numeric(x_vals)
# f_vals = f_lambdified(x_vals)
F_vals = F_lambdified(x_vals)

# Criando o gráfico
plt.figure(figsize=(12, 5))

# Subgráfico 1: área sob a curva
plt.subplot(1, 2, 1)
plt.plot(x_vals, f_vals, label=r'$f(x) = \frac{1}{x}e^{\ln{x}}$', color='orange')
plt.fill_between(x_vals, f_vals, where=(x_vals>=1)&(x_vals<=np.e), alpha=0.3, color='orange', label=r'A = $e-1$ u.a.')
# Limitando visualmente a área real do gráfico
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(np.e, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.title("Integral definida como área")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)

# Subgráfico 2: f(x) e sua integral acumulada
# Definir a primitiva acumulada e seus valores
F_acumulada = sp.integrate(f, (x, 1, x))
F_acumulada_lamb = sp.lambdify(x, F_acumulada, 'numpy')
F_acumulada_vals = F_acumulada_lamb(x_vals)
plt.subplot(1, 2, 2)
plt.plot(x_vals, f_vals, label=r'$f(x) = \frac{1}{x}e^{\ln{x}}$')
plt.plot(x_vals, F_acumulada_vals, label=r'$F(x) = x$', linestyle='--')
# Plotando o ponto (e, F(e))
plt.scatter([np.e], [F_acumulada_lamb(np.e)], color="red", zorder=3, marker='o', label=r"$F(e) = e-1$")
# Limitando visualmente a área real do gráfico
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(1, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.axvline(np.e, color='black', alpha=0.5, linewidth=0.7, linestyle='--')
plt.title("Função e integral acumulada")
plt.legend()
plt.grid(linestyle='--', alpha=0.3)

plt.show()
```
